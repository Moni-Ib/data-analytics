```python
from sklearn.datasets import make_regression
```

```python
X, y= make_regression(n_features=1, noise=30, random_state=0) #randomstate es el estado aleatorio, para que los datos sean lo mismo cada vez
```

```python
X.shape, y.shape
```

```python
import pandas as pd 
df=pd.DataFrame(X)
df.columns = ["x"]
df["y"]=y
df
```

```python
df["x"].plot(kind="hist")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
import matplotlib.pyplot as plt #clase el tipo de modelo

model_1=LinearRegression() #quiero un objeto de tipo linearRegression
model_2=LinearRegression()
model_3=LinearRegression()

model_1.fit(X,y)
```

```python
y_pred = model_1.predict(X)
```

```python
import numpy as np

y_mean=np.array([y.mean()]*100)
```

```python
y.mean()
```

```python
#Selector
plt.figure(figsize=(10,4)) #tamaÃ±o de la grÃ¡fica

#Data
plt.scatter(X,y, label="Random data")

plt.plot(X,y_mean, label="Mean", c="green")

plt.plot(X,y_pred, label="Linear Regression", c="red")

#show 
plt.xlabel("Feature")
plt.ylabel("Target (y)")
plt.title("Scatter example plot")
plt.legend()

plt.savefig("../reports/figures/random1.svg")
```
